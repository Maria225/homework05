```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_to_load)
trial_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
mousetrial_df = pd.merge(mouse_df, trial_df, on="Mouse ID", how="outer")

# Display the data table for preview
mousetrial_df

#mousetrial_df.to_dense().to_csv("mousetrial_output.csv", index = False, sep=',', encoding='utf-8')
mousetrial_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint and Convert to DataFrame
mean_tv_df = mousetrial_df.groupby(["Drug", "Timepoint"]).mean()["Tumor Volume (mm3)"].reset_index()

# Preview DataFrame
mean_tv_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint & Convert to DataFrame
sem_tv_df = mousetrial_df.groupby(["Drug", "Timepoint"]).sem()["Tumor Volume (mm3)"].reset_index()

# Preview DataFrame
sem_tv_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_tv_pivot_df = mean_tv_df.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]

# Preview that Reformatting worked
mean_tv_pivot_df
```

```python
sem_tv_pivot_df= sem_tv_df.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]
sem_tv_pivot_df
```

```python
# Generate the Plot (with Error Bars)
Capomulin = plt.errorbar(mean_tv_pivot_df.index, mean_tv_pivot_df["Capomulin"], yerr=sem_tv_pivot_df["Capomulin"], color="r", marker="o", markersize=5, linestyle="dotted", label="Capomulin")
Infubinol = plt.errorbar(mean_tv_pivot_df.index, mean_tv_pivot_df["Infubinol"], yerr=sem_tv_pivot_df["Infubinol"], color="b", marker="^", markersize=5, linestyle="dotted", label="Infubinol")
Ketapril = plt.errorbar(mean_tv_pivot_df.index, mean_tv_pivot_df["Ketapril"], yerr=sem_tv_pivot_df["Ketapril"], color="g", marker="s", markersize=5, linestyle="dotted", label="Ketapril")
Placebo = plt.errorbar(mean_tv_pivot_df.index, mean_tv_pivot_df["Placebo"], yerr=sem_tv_pivot_df["Placebo"], color="k", marker="d", markersize=5, linestyle="dotted", label="Placebo")

plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mmm3)")
plt.legend(handles=[Capomulin, Infubinol, Ketapril, Placebo], loc="best")
plt.grid(axis='y')

# Save the Figure
plt.savefig("Tumor Volume Changes Over Time")

# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Minor Data Munging to Re-Format the Data Frames
mean_mets_pivot_df = mean_mets_df.pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]

# Preview that Reformatting worked
mean_mets_pivot_df

##this does not look like the answer key, but i think it is still correct, because at time 0, mets was 0 for all the drugs
```

```python
sem_mets_pivot_df= sem_mets_df.pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]
sem_mets_pivot_df
```

```python
# Generate the Plot (with Error Bars)
Capomulin = plt.errorbar(mean_mets_pivot_df.index, mean_mets_pivot_df["Capomulin"], yerr=sem_mets_pivot_df["Capomulin"], color="r", marker="o", markersize=5, linestyle="dotted", label="Capomulin")
Infubinol = plt.errorbar(mean_mets_pivot_df.index, mean_mets_pivot_df["Infubinol"], yerr=sem_mets_pivot_df["Infubinol"], color="b", marker="^", markersize=5, linestyle="dotted", label="Infubinol")
Ketapril = plt.errorbar(mean_mets_pivot_df.index, mean_mets_pivot_df["Ketapril"], yerr=sem_mets_pivot_df["Ketapril"], color="g", marker="s", markersize=5, linestyle="dotted", label="Ketapril")
Placebo = plt.errorbar(mean_mets_pivot_df.index, mean_mets_pivot_df["Placebo"], yerr=sem_mets_pivot_df["Placebo"], color="k", marker="d", markersize=5, linestyle="dotted", label="Placebo")

plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met Sites")
plt.legend(handles=[Capomulin, Infubinol, Ketapril, Placebo], loc="best")
plt.grid(axis='y')

# Save the Figure
plt.savefig("Number Of Metastatic Sites Over Time")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric) & Convert to DataFrame
mice_count_df = mousetrial_df.groupby(["Drug", "Timepoint"]).count()["Mouse ID"].reset_index()
mice_count_df = mice_count_df.rename(columns={"Mouse ID": "Mouse Count"})

# Preview DataFrame
mice_count_df 
```

```python
# Minor Data Munging to Re-Format the Data Frames
mice_count_pivot_df = mice_count_df.pivot(index="Timepoint", columns="Drug")["Mouse Count"]

# Preview the Data Frame
mice_count_pivot_df
```

```python
mice_percent_df = mice_count_pivot_df
mice_percent_df = ((mice_percent_df.loc[:]/mice_percent_df.loc[0])*100)
mice_percent_df
```

```python
# Generate the Plot (Accounting for percentages)
Capomulin, = plt.plot (mice_count_pivot_df.index, mice_percent_df["Capomulin"], color="r", marker="o", markersize=5, linestyle="dotted", label="Capomulin")
Infubinol, = plt.plot(mice_count_pivot_df.index, mice_percent_df["Infubinol"], color="b", marker="^", markersize=5, linestyle="dotted", label="Infubinol")
Ketapril, = plt.plot(mice_count_pivot_df.index, mice_percent_df["Ketapril"], color="g", marker="s", markersize=5, linestyle="dotted", label="Ketapril")
Placebo, = plt.plot(mice_count_pivot_df.index, mice_percent_df["Placebo"], color="k", marker="d", markersize=5, linestyle="dotted", label="Placebo")

plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.legend(handles=[Capomulin, Infubinol, Ketapril, Placebo], loc="best")
plt.grid()

# Save the Figure
plt.savefig("Mice Survival Through Treatment Course")

# Show the Figure
plt.show()
```
